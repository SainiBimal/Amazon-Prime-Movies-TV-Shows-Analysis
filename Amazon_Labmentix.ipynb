```python
title["tmdb_popularity"] = title["tmdb_popularity"].fillna(round(title["tmdb_popularity"].mean(),3))
```

```python
title["tmdb_score"] = title["tmdb_score"].fillna(round(title["tmdb_score"].mean(),1))
```

# **Histogram Plot**

A histogram is a graphical representation that shows the distribution of a numerical variable by dividing the data into continuous intervals or bins.

**Histogram Plot of IMDB Score**

```python
# Histogram Plot of IMDB Score
plt.figure(figsize=(12,6))
sns.histplot(data= data, x="imdb_score" , bins =30, kde =True)
plt.title("Data distribution of Imdb Score")
plt.show()
```

**Insights :**

*   The data distribution of IMDB Score mostly high at 6.0 to 7.0.




**Histogram Plot of Runtime**

```python
# Histogram Plot of Runtime
plt.figure(figsize=(12,6))
sns.histplot(data= data, x="runtime" , bins =30, kde =True)
plt.title("Data distribution of Runtime")
plt.show()
```

**Insights :**

*   The data distribution for runtime is high at 100 minutes, after that there is drop.



**Histogram Plot of TMDB Score**

```python
# Histogram Plot of TMDB Score
plt.figure(figsize=(12,6))
sns.histplot(data= data, x="tmdb_score" , bins =30, kde =True)
plt.title("Data distribution of TMDB Score")
plt.show()
```


**Insights :**


* The scores are approximately normally distributed, centered around 6.0.

# **Box Plot**

A box plot is a graphical representation used to show the distribution, central value, and variability of a dataset. Helpful for comparing distributions across categories and detecting outliers, skewness, and spread in numerical data.

**Box Plot of IMDB Score by Type**

```python
# Box Plot of IMDB Score by Type
plt.figure(figsize=(10,8))
sns.boxplot(data = data, x="type", y="imdb_score", color = "orange")
plt.title("Box plot of Imdb Score by Type")
plt.show()
```

**Insights :**

*   TV Shows have higher IMDb scores than Movies, with a higher median and tighter interquartile range.

*  Movies has a wider range of ratings and more outliers.

*  TV Shows are more consistently rated, while Movies include many low-rated titles and a few extremely high-rated ones.


**Box plot of Tmdb Score by Type**

```python
# Box plot of Tmdb Score by Type
plt.figure(figsize=(10,8))
sns.boxplot(data = data, x="type", y="tmdb_score", color = "orange")
plt.title("Box plot of Tmdb Score by Type")
plt.show()
```

**Insights :**

* TV Shows have higher TMDb scores on average than Movies, with a median score
closer to 7.5 compared to Movies around 6.

* Movies exhibit greater variability with more extreme outliers on the lower end.

* TV Shows are more consistently rated than the Movie.

**Box plot of Runtime by Type**

```python
# Box plot of Runtime by Type
plt.figure(figsize=(10,6))
sns.boxplot(data = data, x="type", y="runtime", color = "red")
plt.title("Box plot of Runtime by Type")
plt.show()
```


**Insights :**


* The median runtime for Movies is significantly higher 100 mins than for Shows 45 mins.

# **Line Chart**

A line chart is a type of graph that displays data points connected by straight lines. It is commonly used to show trends over time or to visualize changes in continuous data.

**Line plot of Runtime by Type based on TMDB Popularity**

```python
# Line plot of Runtime by Type based on TMDB Popularity
plt.figure(figsize=(12,6))
sns.lineplot(data = run_tmdb, x="runtime", y ="tmdb_popularity", hue="type", palette = "tab10")
plt.title("Movies and Shows type's Runtime based on TMDB Popularity ")
plt.xticks(rotation =90)
plt.show()
```

**Insights :**

# **PIE Chart**

A pie chart is a circular statistical graphic that is divided into slices to illustrate proportions or percentages of a whole. Each slice of the pie represents a category's contribution to the total, making it easy to compare parts to the whole.

# **Multivariate Analysis**

Multivariate analysis refers to statistical techniques used to analyze more than two variables simultaneously to understand relationships, patterns, or impacts among them.

# **Correlation Heatmap**

A correlation heatmap is a visual representation of the correlation matrix showing how strongly variables relate to each other.

```python
data.dtypes
```

```python
# Correlation Matrix
numb = data.select_dtypes(include=["int64", "float64"])
corr1 = numb.corr()
corr1
```

**Correlation Heatmap Chart**

```python
# Correlation Heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(corr1, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

**Insights :**

* IMDb Score and TMDb Score have a strong positive correlation of 0.62, indicating that content rated highly on IMDb tends to also receive high scores on TMDb.

* Runtime and IMDb Votes show a moderate positive correlation of 0.22, suggesting that longer content may attract more viewer attention and votes.

* Seasons and Runtime have a negative correlation of -0.30, implying that shows with more seasons tend to have shorter average runtimes per episode.

# **Pair Plots**
