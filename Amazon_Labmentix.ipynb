# **Project Summary**

In Amazon Prime TV Shows and Movies analysis project the task is given to perform Exploratory Data Analysis (EDA) based on  number of content, trends, audience preferences and drive the useful insights from them as per the business strategy.

To start with project here are few steps-

# 1) Understand the Problem Statement -

* What types of content (movies or shows) dominate the platform.

* Which genres are most popular.

* How ratings, runtimes, or content trends changed over the years.

* Which countries or creators contribute the most.

* Find top movies, actors, directors.



# 2) Data Collection -

 We have given two datasets

* **Titles Dataset (titles.csv)**


**id** -	Title ID on JustWatch

**title**	- Name of the TV show or movie

**show_type** - 	Whether it is a TV show or movie

**description** - 	Brief content description

**release_year**	 - Year of release

**age_certification**	 - Official age rating

**runtime**	- Duration of the movie or episode

**genres**	- List of genres associated

**production_countries** -	Countries that produced the content

**seasons** - 	Number of seasons (for TV shows only)

**imdb_id**	- Unique IMDb identifier

**imdb_score**	- IMDb rating score

**imdb_votes**	- Number of IMDb votes

**tmdb_popularity**- 	Popularity score on TMDB

**tmdb_score** - 	Rating score on TMDB



* **Credits Dataset (credits.csv)**

**person_id** - Person ID on JustWatch

**id**	- Corresponding title ID (linked to titles.csv)

**name**	-  Actor or director's name

**character_name**	-  Name of the character portrayed

**role**	-  Whether the person is an ACTOR or DIRECTOR

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
credit = pd.read_csv("/content/drive/MyDrive/Labmentix Internship 2025/credits.csv")
title = pd.read_csv("/content/drive/MyDrive/Labmentix Internship 2025/titles.csv")
```

```python
credit["character"] = credit["character"].fillna("Unknown")
credit.isnull().sum()
```

```python
credit.duplicated().sum()
```

```python
credit = credit.drop_duplicates()
```

```python
credit.info()
```

```python
title
```

```python
title.isnull().sum()
```

```python
title["description"] = title["description"].fillna("Unknown Description")
```

```python
title["age_certification"] = title["age_certification"].fillna("Not Rated")
```

```python
title["seasons"] = title["seasons"].fillna(0)
```

```python
title.dropna(subset = ["imdb_id"], inplace = True)
```

```python
title["imdb_score"] = title["imdb_score"].fillna(round(title["imdb_score"].mean(),1))
```

```python
title["imdb_votes"] = title["imdb_votes"].fillna(round(title["imdb_votes"].mean(),1))
```

```python
title["tmdb_popularity"] = title["tmdb_popularity"].fillna(round(title["tmdb_popularity"].mean(),3))
```

```python
title["tmdb_score"] = title["tmdb_score"].fillna(round(title["tmdb_score"].mean(),1))
```

```python
title.isnull().sum()
```

```python
title.dtypes
```

```python
title.duplicated().sum()
```

```python
title = title.drop_duplicates()
title.duplicated().sum()
```

```python
title.info()
```

```python
data = pd.merge(title, credit, on = "id", how = "inner")
data
```

```python
plt.figure(figsize=(12,6))
sns.histplot(data= data, x="imdb_score" , bins =30, kde =True)
plt.title("Data distribution of Imdb Score")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data = data, x="type", y="tmdb_score", color = "orange")
plt.title("Box plot of Tmdb Score by Type")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data = data, x="type", y="runtime", color = "red")
plt.title("Box plot of Runtime by Type")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.histplot(data= data, x="tmdb_score" , bins =30, kde =True)
plt.title("Data distribution of TMDB Score")
plt.show()
```

```python
run_tmdb =data.groupby(["runtime","type"]).agg({"tmdb_popularity" : "sum"}).sort_values(by = "tmdb_popularity", ascending = False)
run_tmdb
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data = run_tmdb, x="runtime", y ="tmdb_popularity", hue="type", palette = "tab10")
plt.title("Movies and Shows type's Runtime based on TMDB Popularity ")
plt.xticks(rotation =90)
plt.show()
```

```python
year_type = data.groupby(["release_year" , "type"]).agg({"type" : "count"}).rename(columns = {"type" : "No of Movies and Shows"}).reset_index()
year_type
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data= year_type, x = "release_year", y= "No of Movies and Shows",  hue = "type" , palette = "tab10")
plt.title("Year-wise Movies and Shows comparison")
plt.show()
```

```python
imdb_type = data.groupby(["imdb_score" , "type"]).agg({"type" : "count"}).rename(columns = {"type" : "No of Movies and Shows"}).reset_index()
imdb_type
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data= imdb_type, x = "imdb_score", y= "No of Movies and Shows",  hue = "type" , palette = "tab10")
plt.title("IMDB Score comparison for Movies and Shows.")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(data= data,  x= "release_year", y= "runtime",  hue = "type", palette = "tab10")
plt.title("Release Year wise Runtime comparision for Movies and Shows")
plt.show()
```

```python
prod = data.groupby(["production_countries","type"]).agg({"production_countries" : "count"}).rename(columns={"production_countries" : "count"}).sort_values(by = "count", ascending = False ).reset_index()
prod.head(10)
```

```python
age_cert = data.groupby(["age_certification", "type"]).agg({"age_certification":"count"}).rename(columns={"age_certification" : "count"}).sort_values(by = "count", ascending =False ).reset_index()
age_cert
```

```python
plt.figure(figsize=(18,6))
sns.barplot(data= age_cert,  x= "age_certification", y = "count",  hue = "type", palette = "tab10")
plt.title("Age certification wise comparision for Movies and Shows")
plt.show()
```

```python
genres_imdbvote = data.groupby(["genres"]).agg({"imdb_votes" : "sum"}).sort_values(by = "imdb_votes", ascending = False).head(20)
genres_imdbvote
```

```python
plt.figure(figsize=(18,6))
sns.barplot(data= genres_imdbvote, x= "genres", y = "imdb_votes", color = "pink")
plt.title("Top 20 Genres based on Imdb Votes")
plt.xticks(rotation =90)
plt.ticklabel_format(style="plain", axis="y")
plt.show()
```

```python
Top_movie_show = data.groupby(["title", "type"]).agg({"imdb_score" : "sum"}).sort_values(by = "imdb_score", ascending = False).head(30)
Top_movie_show
```

```python
director = data[data["role"]=="DIRECTOR"]
director = director.groupby(["name","type"]).agg({"imdb_votes" : "sum"}).sort_values(by = "imdb_votes", ascending = False).head(30)
director
```
